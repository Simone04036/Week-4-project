```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup

URL = 'https://www.helmet.beam.vt.edu/bicycle-helmet-ratings.html#!'

# Setting up the Selenium driver (assuming chromedriver is in PATH)
driver = webdriver.Chrome()

# Navigate to the page
driver.get(URL)

# Let's give it some time to load the dynamic content
driver.implicitly_wait(10)  # waits up to 10 seconds for elements to appear

# Parse the loaded page with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find the specific container
portfolio_div = soup.find('div', id='portfolio', class_='section gray')

brands = []
product_names = []
prices = []
scores = []
star_ratings = []

if portfolio_div:
    gallery_headers = portfolio_div.find_all('div', class_='gallery-header gallery-header-bottom')
    helmet_names = portfolio_div.find_all('span', class_='helmet-name')
    
    for name_div, header in zip(helmet_names, gallery_headers):
        full_name = name_div.b.text
        brand = full_name.split()[0]
        product_name = " ".join(full_name.split()[1:])
        
        spans = header.find_all('span')
        price = spans[0].text.split(": ")[1]
        score = spans[1].text.split(": ")[1]
        
        star_div = header.find('span', class_='right-gallery-rating')
        total_stars = 5
        empty_stars = star_div.i.text.count("star_border")
        filled_stars = total_stars - empty_stars

        brands.append(brand)
        product_names.append(product_name)
        prices.append(price)
        scores.append(float(score))  # Converted to float for calculations
        star_ratings.append(filled_stars)

    data = {
        'Brand': brands,
        'Product Name': product_names,
        'Price': prices,
        'Safety Score': scores,
        'Star Rating': star_ratings
    }

    df = pd.DataFrame(data)
    print(df)

else:
    print("Could not find the desired container")

# Close the browser window
driver.quit()  # Corrected to use quit() instead of Quit()
```

Test hypotheses "Lazer brand helmets, on average, have a safety score that's significantly better than the industry average."

```python
import scipy.stats as stats

# Sample data: Safety scores of Lazer brand helmets
lazer_scores = df[df['Brand'] == 'Lazer']['Safety Score']

# Population mean: Industry average safety score
industry_average = df['Safety Score'].mean()

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(lazer_scores, industry_average)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision based on p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There's significant evidence that the average safety score for Lazer brand helmets is different from the industry average.")
else:
    print("Fail to reject the null hypothesis. There isn't enough evidence to suggest that the average safety score for Lazer brand helmets is different from the industry average.")

```

Note:
If the t-statistic is positive, it indicates that the sample mean (Lazer brand's average safety score) is greater than the population mean (industry average), which would not be desirable since a lower safety score is better. Conversely, a negative t-statistic would indicate that Lazer's average is better than the industry's.
A small p-value (typically ≤ 0.05) would indicate that there's strong evidence against the null hypothesis and you would reject it in favor of the alternative hypothesis.
This test assumes that the safety scores are approximately normally distributed. If that assumption doesn't hold, the results might not be reliable. Consider checking the distribution of the safety scores before performing the t-test.
